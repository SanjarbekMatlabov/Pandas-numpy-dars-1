```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
import sklearn
df_ca = {
    'ism' : ['Ali', 'Vali', 'Bali'],
    'yosh' : [23,22,20],
    'oylik': [300,500,700],
    'boyi' : [180.664,190.564,170.765],
    'hobbi': ['futbol', 'voleybol', 'tenis'],
    'positsiya' : ['student', 'ishchi', 'bekorchi'],
    'rating' : ['A', 'B', 'C'],
    'status': ['yolgiz', 'farzandli', 'uylangan']
}
df = pd.DataFrame(df_ca)
df
```

```python
dummies = pd.get_dummies(df,columns=["hobbi","positsiya"],dtype=np.int8)
dummies
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder(dtype=np.int8).set_output(transform='pandas')
df_ordinal = ordinal.fit_transform(dummies[['rating','status']])
df_ordinal
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
onehot = OneHotEncoder(dtype=np.int8,sparse_output=False).set_output(transform='pandas')
ho_ordinal = onehot.fit_transform(df[['hobbi','positsiya']])
ho_ordinal  
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
import sklearn
titanic = sns.load_dataset("titanic")
titanic
```

```python

a = titanic.isnull().sum()
a
```

```python

titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic

```

```python
titanic['embarked'] = titanic['embarked'].astype('category')
titanic['embarked']
```

```python
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic.isnull().sum()
```

```python
titanic['deck'] = titanic['deck'].fillna(titanic['deck'].mode()[0])
titanic.isnull().sum()
```

```python
titanic
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Unknown')
titanic
```

```python
titanic = sns.load_dataset("titanic")

selected_columns = ['survived', 'sex', 'embarked', 'class', 'age', 'fare']
titanic = titanic[selected_columns]

titanic
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

titanic
```

```python
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic

```

```python
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic
```

```python
titanic = pd.get_dummies(titanic, columns=['embarked'], drop_first=True)
titanic

```

```python
titanic = pd.get_dummies(titanic, columns=['class'])
titanic

```

```python
titanic['fare'] = titanic['fare'].replace(0, titanic['fare'].median())
titanic
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic[['age', 'fare']] = scaler.fit_transform(titanic[['age', 'fare']])
titanic
```
